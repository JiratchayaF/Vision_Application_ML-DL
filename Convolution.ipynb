```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
#import image
image = cv2.imread('/content/apple.jpg',0)
image.shape
```

```python
filter = np.array([[-1, 0, 1],[-2, 0, 2],[-1, 0, 1]])
filter
```

```python
width = image.shape[0]
height = image.shape[1]
print(width,height)
```

```python
#padding
padding = np.zeros((width + 2, height + 2))
padding.shape
padding[1:width+1,1:height+1] = image
plt.imshow(padding)
```

```python
input = np.zeros(image.shape) #####
sum1 = np.zeros(([width,height]))
sum1
```

```python
for r in range(height):
  for c in range(width):
    input = padding[c:c+3,r:r+3]
    sum1[c,r] = np.sum(filter * input)
print(sum1)
cv2_imshow(sum1)  #####
```
